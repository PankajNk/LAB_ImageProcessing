```python
import matplotlib.pyplot as plt
import cv2 
import numpy as np

```

```python
img = cv2.imread("./image/flower.jpg",0)
```

```python
#histogram 
plt.subplot(1,2,1)
plt.imshow(img,cmap ="gray")

plt.subplot(1,2,2)
plt.hist(img.ravel(),256,[0,256])

plt.show()
```

```python
eq = cv2.equalizeHist(img)

plt.subplot(1,2,1)
plt.imshow(img,cmap ="gray")

plt.subplot(1,2,2)
plt.hist(eq.ravel(),256,[0,256])

plt.show()
```
