```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("./image/Tiger.jpg")
#https://learnopencv.com/edge-detection-using-opencv/
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#display both
```

```python
#blurring is reqd bcs its 1st order...theory
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)
#cv2.imshow(" Gaussican Blur",img_blur)
#k=cv2.waitKey(0)
#if k == 27:
#    cv2.destroyAllWindows()
```

```python
sobelx = cv2.Sobel(src = img_blur,ddepth=cv2.CV_64F, dx=1 , dy=0 ,ksize=5) # Sobel Edge

sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge

sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X 
```

```python
#display all 
cv2.imshow('Sobel X', sobelx)
k=cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
```

```python
cv2.imshow('Sobel Y', sobely)
k=cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
```

```python
cv2.imshow('Sobel X Y using Sobel() function', sobelxy)
k=cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
```

```python
#...theory
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200) # Canny Edge Detection

# Display Canny Edge Detection Image
cv2.imshow('Canny Edge Detection', edges)
k=cv2.waitKey(0)
if k == 27:
    cv2.destroyAllWindows()
```
